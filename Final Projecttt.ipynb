# Preprocessing

- Are there any duplicated values?
- Do we need to do feature scaling?
- Do we need to generate new features?
- Split Train and Test dataset. (0.7/0.3)

```python
# Cleaning & Looking for duplicate values
```

```python
d.drop_duplicates()
```

```python
# Looking for outliers with z-score method

z = np.abs(stats.zscore(d))
z
```

```python
outliers = list(set(np.where(z > 3)[0]))
len(outliers)
# we got 148 outliers
```

```python
new_df = d.drop(outliers, axis = 0).reset_index(drop=False)
new_df.head(6)
```

```python
# Starting with model

from sklearn import preprocessing
X = d.drop(['fixed acidity'], axis = 1) 
y = d[['fixed acidity']]



# scaling
X_s = preprocessing.scale(X)
X_s = pd.DataFrame(X_s, columns = X.columns)

y_s = preprocessing.scale(y)
y_s = pd.DataFrame(y_s, columns = y.columns)

X_s
```

```python
label_encoder = LabelEncoder()
d["Label"] = label_encoder.fit_transform(d["quality"])
d.head()
```

```python
categories = list(set(d.quality))
d.drop(["quality"], axis=1, inplace=True)
d.head()
```

```python
#Locating features and label
X,y = d.iloc[:,:-1], d.iloc[:,-1]

#Splitting dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
X_test
```

# ML Application

- Define models.
- Fit models.
- Evaluate models for both train and test dataset.
- Generate Confusion Matrix and scores of Accuracy, Recall, Precision and F1-Score.
- Analyse occurrence of overfitting and underfitting. If there is any of them, try to overcome it within a different section.

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=17, random_state= 18)
clf.fit(X_train, y_train)
print("Accuracy of train:",clf.score(X_train, y_train))
print("Accurayc of test:", clf.score(X_test, y_test))
print("Bias for training:",1-clf.score(X_train,y_train))
print("Bias for test:",1-clf.score(X_test,y_test))
print("Variance:",(1-clf.score(X_test,y_test))-(1-clf.score(X_train,y_train)))
```

```python
# Feature Importances

plt.figure(figsize=(12, 4))
importance = clf.feature_importances_
sns.barplot(x=importance, y=X.columns)
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score, classification_report, f1_score
pred = clf.predict(X_test)
print(classification_report(y_test,pred))
```

```python
# CONFUSION MATRIX

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, pred)
plt.figure(figsize=(12, 8))
ax =sns.heatmap(cm, square=True, annot=True, cbar=False)
ax.xaxis.set_ticklabels(categories, fontsize = 12)
ax.yaxis.set_ticklabels(categories, fontsize = 12, rotation=0)
ax.set_xlabel('Predicted Labels',fontsize = 15)
ax.set_ylabel('True Labels',fontsize = 15)
plt.show()
```

```python
# XGBOOST ALGO

import xgboost as xgb

dmatrix_train = xgb.DMatrix(data=X_train, label=y_train)
dmatrix_test = xgb.DMatrix(data=X_test, label=y_test)

param = {'max_depth':17, 
         'eta':1, 
         'objective':'multi:softprob', 
         'eval_metric':'mlogloss',
         'num_class':6}

num_round = 6
model = xgb.train(param, dmatrix_train, num_round)

categories
```

```python
# PREDIC

preds = model.predict(dmatrix_test)
preds[:10]
```

```python
# BEST PREDICTIONS

best_preds = np.asarray([np.argmax(line) for line in preds])
best_preds
```

```python
print("Precison = {}".format(precision_score(y_test, best_preds, average='macro')))
print("Recall = {}".format(recall_score(y_test, best_preds, average='macro')))
print("Accurcy = {}".format(accuracy_score(y_test, best_preds)))
```
